```python
# Count how many SalePrice values are >= 500,000
count_over_500k = (df['SalePrice'] >= 500000).sum()

# Count how many SalePrice values are >= 500,000
count_over_400k = (df['SalePrice'] >= 400000).sum()

# Total number of non-missing SalePrice values
total = df['SalePrice'].notna().sum()

# Percentage
percentage_over_500k = (count_over_500k / total) * 100

percentage_over_400k = (count_over_400k / total) * 100

count_over_500k, total, percentage_over_500k, percentage_over_400k
```

```python
# Define bins and labels
bins = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000]
labels = [
    '0100k',
    '100k200k',
    '200k300k',
    '300k400k',
    '400k500k',
    '500k600k',
    '600k700k'
]

# Create categorical bins
price_bins = pd.cut(df['SalePrice'], bins=bins, labels=labels, right=False)

# Calculate percentage distribution
percentage_distribution = (
    price_bins.value_counts(normalize=True)
    .sort_index()
    * 100
)

# Display results
percentage_distribution

import matplotlib.pyplot as plt

# Plot histogram (counts per price range)
price_bins.value_counts().sort_index().plot(kind='hist')

plt.xlabel("Sale Price Range")
plt.ylabel("Number of Houses")
plt.title("Histogram of House Sale Prices by Range")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Single bars, clear separation
plt.hist(
    df['SalePrice'],
    bins=30,            # number of bins (one bar per range)
    edgecolor='black',  # outline each bar
    linewidth=1,
    rwidth=1.0          # full width per bin, no overlap
)

plt.xlabel("Sale Price")
plt.ylabel("Number of Houses")
plt.title("Histogram of House Sale Prices")
plt.grid(axis='y', alpha=0.3)  # horizontal grid for clarity
plt.tight_layout()
plt.show()
```
